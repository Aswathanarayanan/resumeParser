```python
import requests

def download_file_from_drive(drive_link, output_dir,file_name):
    # Extract file ID from the Google Drive link
    file_id = drive_link.split('/')[-1]
    file_id = file_id.split('=')[-1]
    print(file_id)

    download_url=drive_link.replace("open","uc")
    print(download_url)
    # Construct download URL
    # download_url = "https://drive.google.com/uc?id={file_id}"
    file_name="/content/resume/"+file_name+".pdf"
    # Get file name from the end of the URL
    # file_name = drive_link.split('/')[-1]
    # Download the file
    response = requests.get(download_url)
    if response.status_code == 200:
        with open(file_name, 'wb') as f:
            f.write(response.content)
        print(f"File '{file_name}' downloaded successfully.")
    else:
        print("Failed to download the file.")
```

```python
import pytesseract

def extract_text_from_image(image):
    text = pytesseract.image_to_string(image)
    return text
```

```python
import fitz
import pytesseract
from PIL import Image

def pdf_to_text_with_ocr(pdf_path):
    # Open the PDF file
    try:
      doc = fitz.open(pdf_path)

      # Initialize empty text
      text = ''

      # Iterate over each page in the PDF
      for page_num in range(len(doc)):
          # Get the page
          page = doc.load_page(page_num)

          # Convert the page to an image
          image = page.get_pixmap()

          # Convert the image to grayscale
          image = Image.frombytes("RGB", [image.width, image.height], image.samples)

          # Perform OCR on the image
          page_text = pytesseract.image_to_string(image)

          # Append the extracted text to the result
          text += page_text

      return text

    except Exception as e:
      print("error in opening the file",pdf_path)
```

```python
from fuzzywuzzy import fuzz
def extract_skills(doc, skills_list):
    skills = []
    for ent in doc.ents:
      t=ent.text.lower()
      for skill in skills_list:
        similarity_score = fuzz.ratio(skill, t)
        if similarity_score > 80:
            skills.append(t)
    return skills
```

```python
def extract_years_from_text(text):
    # Define regex pattern to match years (e.g., 2022, 1998)
    pattern = r'\b(?:19|20)\d{2}\b'
    years = re.findall(pattern, text)
    return years
```
